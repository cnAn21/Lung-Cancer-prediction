#**Data visualization**

Bar plot :

```python
#Visualizing Lung cancer prevalance by Gender
gc = df.groupby('GENDER')['LUNG_CANCER'].value_counts()


gc = gc.unstack()

genders = gc.index
no_cancer_counts = gc[0]
yes_cancer_counts = gc[1]


plt.bar(genders, no_cancer_counts, width=0.4, label='No Lung Cancer', color='y')
plt.bar(genders+0.4, yes_cancer_counts, width=0.4, label='Lung Cancer', color='r')

plt.xlabel("Gender")
plt.ylabel("Lung Cancer Count")
plt.title("Lung Cancer Prevalence by Gender")
plt.xticks(genders)
plt.legend()
plt.show()
```

  This visualization indicates that males have a higher count of lung cancer cases compared to females.

Histogram :

```python
#Visualizing age distribution for Lung cancer
no_cancer_df = df[df['LUNG_CANCER'] == 0]
yes_cancer_df = df[df['LUNG_CANCER'] == 1]


plt.hist(no_cancer_df['AGE'], bins=10, alpha=1.0, label='No Lung Cancer', color='blue')
plt.hist(yes_cancer_df['AGE'], bins=10, alpha=0.5, label='Lung Cancer', color='red')

plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution for Lung Cancer Groups")
plt.legend()
plt.show()
```

Based on this visualization, we can see that lung cancer is more common in older adults. The frequency of lung cancer increases steadily starting from around 40 years old, and it is highest in people in their 70s and 80s.

Heatmap

```python
# Investigating Correlations with Lung Cancer
cause_columns = ['SMOKING', 'YELLOW_FINGERS', 'ANXIETY','PEER_PRESSURE', 'CHRONIC DISEASE','FATIGUE ', 'ALLERGY ', 'WHEEZING','ALCOHOL CONSUMING', 'COUGHING','SHORTNESS OF BREATH','SWALLOWING DIFFICULTY', 'CHEST PAIN']
relevant_columns = cause_columns + ['LUNG_CANCER']


correlation_matrix = df[relevant_columns].corr()



sns.heatmap(correlation_matrix)
plt.title("Correlation with Lung Cancer")
plt.show()
```

This visualization highlights a strong correlation between smoking and lung cancer.  Other factors like yellow fingers, anxiety, and chronic cough also appear connected to lung cancer, although to a lesser degree.

#**Conclusion:**

Overall the data suggests a higher prevalence of lung cancer in males compared to females, with a tendency for it to occur more frequently in older adults.Smoking appears to be a strong risk factor, while other factors like anxiety and chronic cough may also be associated with lung cancer.
