#**INTRODUCTION**

This dataset is related to respiratory health and potential risk factors for lung cancer.

   It includes information on:
* **Demographics :** Gender and Age
* **Habits and Lifestyle:** Smoking, Alcohol Consumption
* **Symptoms :** Fatigue, Coughing, Shortness of Breath, Chest Pain
* **Pre-existing Conditions :** Chronic Disease, Allergy
* **Potential Risk Factors :** Yellow Fingers, Anxiety, Peer Pressure, Wheezing, Swallowing Difficulty

```python
# Investigating Correlations with Lung Cancer
cause_columns = ['SMOKING', 'YELLOW_FINGERS', 'ANXIETY','PEER_PRESSURE', 'CHRONIC DISEASE','FATIGUE ', 'ALLERGY ', 'WHEEZING','ALCOHOL CONSUMING', 'COUGHING','SHORTNESS OF BREATH','SWALLOWING DIFFICULTY', 'CHEST PAIN']
relevant_columns = cause_columns + ['LUNG_CANCER']


correlation_matrix = df[relevant_columns].corr()



sns.heatmap(correlation_matrix)
plt.title("Correlation with Lung Cancer")
plt.show()
```

```python
#train test split
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
ytrain.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE()
```

```python
xres,yres = sm.fit_resample(xtrain,ytrain)
```

```python
yres.value_counts()
```

##Scale-down

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaler.fit(xtrain)
xtrain = scaler.transform(xtrain)

scaler.fit(xtest)
xtest = scaler.transform(xtest)

scaler.fit(xres)
xres = scaler.transform(xres)
```
